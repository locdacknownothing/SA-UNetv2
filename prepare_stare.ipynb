```python
test_image_files = [file for file in all_image_files if file.stem in test_image_stems]
test_label_files = [file for file in all_label_files if file.stem in test_label_stems]

train_image_files = [file for file in all_image_files if file.stem not in test_image_stems]
train_label_files = [file for file in all_label_files if file.stem not in test_label_stems]

len(train_image_files), len(train_label_files), len(test_image_files), len(test_label_files)
```

```python
train_image_dir = "STARE/train/images"
os.makedirs(train_image_dir, exist_ok=True)

for file in train_image_files:
    image = Image.open(file)

    filename = Path(file).stem + ".png"
    image.save(Path(train_image_dir) / filename)

train_label_dir = "STARE/train/labels"
os.makedirs(train_label_dir, exist_ok=True)

for file in train_label_files:
    label = Image.open(file)

    filename = Path(file).stem + ".png"
    label.save(Path(train_label_dir) / filename)
```
